```python
y0 = 2003
yn = 2018

v0 = 46
vn = 73
```

```python
r_int = 0.07
r_exc = (vn-v0) * 1.0/(yn-y0) *1.0/v0
```

```python
a0 = 1000

n_years = 15
```

```python
A_int = []
A_exc = []

for n in range(n_years):
    
    # Int
    
    an = a0*(1+r_int)**(n)
    A_int += [an]
    
    #Exc
    
    an = a0*(1+r_exc )**(n)
    A_exc += [an]
```
